```python
data_missing_age_value = raw_titanic[raw_titanic['Age'].isnull()]
# 177 rows
```

```python
data_wo_missing_values = raw_titanic.dropna()
# 183 rows 
```

```python
Y = data_wo_missing_values['Age']
```

```python
data_wo_missing_values.columns
```

```python
X = data_wo_missing_values[['Survived', 'Pclass', 'Sex']]
```

```python
from sklearn.utils import shuffle
X, Y = shuffle(X, Y)
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
```

```python
from sklearn.compose import ColumnTransformer

cat_attribs = ['Survived', 'Pclass', 'Sex']

full_pipeline = ColumnTransformer([
        ("cat", OneHotEncoder(), cat_attribs),
    ])

X_hot_catgories = full_pipeline.fit_transform(X)
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg1 = DecisionTreeRegressor(random_state=42, max_depth=5)
tree_reg1.fit(X_hot_catgories, Y)
```

```python
X_for_predict = full_pipeline.fit_transform(data_missing_age_value[['Survived', 'Pclass', 'Sex']])
y_pred1 = tree_reg1.predict(X_for_predict)
```

```python
y_pred1
```

Here we are! All the ages prediction using Decision Tree Model

# Combining lables and values

```python
indics = data_missing_age_value[['Survived', 'Pclass', 'Sex']].index
```

```python
age_with_imputed_values = pd.DataFrame(data =y_pred1, index = indics, columns=  ['Age'])
age_with_imputed_values
```

Cool! now we can merge imputed ages with the related data ( based on indices)

```python
imputed_age_data = pd.concat([age_with_imputed_values, data_missing_age_value.drop(['Age'],axis=1)], axis=1)
imputed_age_data.head()
```
